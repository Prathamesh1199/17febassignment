```python
1: MongoDB is a type of database that is designed to handle large amounts
of data and store it in a way that is flexible and scalable. It does not
use the traditional tables and rows approach of SQL databases, but instead
uses JSON-like documents. Non-relational databases like MongoDB are preferred
when dealing with highly unstructured or dynamic data, or when there is a need 
for horizontal scaling or flexible data modeling.


Non-relational databases, also known as NoSQL databases, are a type of database 
system that do not use a traditional relational data model. Instead of using 
tables and rows to organize and store data, NoSQL databases use a variety of
data models, such as key-value, document-oriented, column-family, and graph
databases, to store and manage data. Non-relational databases are often preferred
in scenarios where data is highly unstructured or dynamic, or when there is
a need for horizontal scaling or flexible data modeling.


MongoDB is often preferred over SQL databases in scenarios where there is
a need for highly flexible data modeling, such as for web applications with
rapidly changing data models. It can also be a good choice for storing 
large amounts of unstructured or semi-structured data, such as social media
posts or sensor data. Additionally, MongoDB is often preferred in scenarios 
where there is a need for horizontal scaling and distributed data storage.



```

```python
2:MongoDB is a popular NoSQL database management system that provides a
number of features that make it a powerful tool for handling large and 
complex datasets. Some of the key features of MongoDB include:
    
1.Document-oriented: MongoDB stores data in a JSON-like format that is
easy to work with.
2.Dynamic schema: Data can be added or modified without having to update
the entire schema.
3.Horizontal scaling: MongoDB can scale horizontally across multiple servers,
allowing for efficient distribution of data and improved performance.
4.Indexing: MongoDB supports various types of indexes, allowing for efficient
querying and retrieval of data
5.Aggregation framework: Allows for complex data analysis and manipulation, 
making it useful for business intelligence and reporting.
6.ACID transactions: MongoDB supports ACID transactions, ensuring data
integrity and consistency.
7.Security: Provides various security features, such as authentication,
authorization, and encryption, to protect data from unauthorized access
and ensure data privacy.
 
    
    
```

```python
3:Here is a code to connect to MongoDB using Python and create a database
  and a collection:
        
# Import the required libraries
from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')

# Create a new database
db = client['mydatabase']

# Create a new collection
collection = db['mycollection']

# Insert a document into the collection
post = {"title": "My first post", "content": "My name is Prathamesh"}
collection.insert_one(post)

# Print the document ID
print("Document ID:", post["_id"])
        
        
In this example, Iam using the pymongo library to connect to the MongoDB server
running on the local machine at the default port 27017.
then i create a new database called mydatabase and a new collection called mycollection.
Finally, we insert a new document into the collection and print its ID.        
        
        
        
        
        
```

```python
4:Here is a code to insert one record and multiple records into the mycollection 
collection, and then print them using the find() and find_one() methods:
    
# Import the required libraries
from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['mydatabase']
collection = db['mycollection']

# Insert one record
post = {"title": "My second post", "content": "This is my second post on MongoDb"}
inserted_post_id = collection.insert_one(post).inserted_id
print("Inserted post ID:", inserted_post_id)

# Insert multiple records
new_posts = [
    {"title": "Post 3", "content": "This is my third post on MongoDB"},
    {"title": "Post 4", "content": "This is my fourth post on MongoDB"}
]
inserted_posts_ids = collection.insert_many(new_posts).inserted_ids
print("Inserted posts IDs:", inserted_posts_ids)

# Print the inserted records
print("\nAll records:")
for post in collection.find():
    print(post)

print("\nOne record:")
print(collection.find_one({"_id": inserted_post_id}))
    
    
   
In this example, i have first connect to the MongoDB server and access the mydatabase
database and mycollection collection.
then i insert one record into the collection and print its ID using the insert_one()
method. Next, i have insert multiple records using the insert_many() method and print their IDs.
 Finally, we use the find() and find_one() methods to print all records and one record, respectively.
 
    
```

```python
5:The find() method is used to query the MongoDB database to retrieve documents that
match a specified query criteria.
The syntax for the find() method is as follows:
    
 db.collection.find(query, projection)
   
    
Where query is the criteria used to filter documents and projection is used to select
which fields to include or exclude from the result set.    
    
Here is a code that demonstrates how to use the find() method to query the
mycollection collection in the mydatabase database for documents with a title field 
containing the word "post":    
    
# Import the required libraries
from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['mydatabase']
collection = db['mycollection']

# Query the collection using the find() method
query = {"title": {"$regex": "post"}}
result = collection.find(query)

# Print the result set
for document in result:
    print(document)
    
    
In this example, I have connect to the MongoDB server and access the mydatabase 
database and mycollection collection.
then i query the collection using the find() method with a query criteria that matches
documents with a title field containing the word "post". 
Finally, we print the resulting documents.   
Iam using the "$regex" operator to perform a regular expression search. This operator
allows us to perform pattern matching and filtering within a field.    
    
    
    
```

```python
6:The sort() method is used to sort the result set of a MongoDB query based on
 one or more fields.
    The syntax for the sort() method is as follows:

db.collection.find(query).sort(sort_key_or_list, direction)

Where sort_key_or_list is the field or fields to sort on and direction is the sorting
order, which can be either ascending (pymongo.ASCENDING) or descending (pymongo.DESCENDING).    
 
    Here an example that demonstrates how to use the sort() method in MongoDB.
Suppose we have a collection called students that contains documents representing
students with fields like name, age, and grade. To sort the collection based on the
grade field in descending order, we can use the following code:
    
# Import the required libraries
from pymongo import MongoClient, DESCENDING

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['mydatabase']
collection = db['students']

# Sort the collection based on the "grade" field in descending order
result = collection.find().sort("grade", DESCENDING)

# Print the sorted result set
for document in result:
    print(document)
    
    
   In this example,I have first connect to the MongoDB server and access the mydatabase 
database and students collection. We then use the sort() method to sort the collection
based on the grade field in descending order.
Finally, we print the sorted result set. 
    
    
```

```python
7:In MongoDB, delete_one(), delete_many(), and drop() are methods used to remove
documents and collections from a database. Here's a brief explanation of each:

1.delete_one() - This method is used to delete a single document that matches a specified 
                 filter. If multiple documents match the filter, only the first one 
                 will be deleted.
2.delete_many() - This method is used to delete all documents that match a specified filter.
3.drop() - This method is used to delete an entire collection from the database.

  These methods are useful for managing the data in a MongoDB database. For example,
we might use delete_one() or delete_many() to remove documents that are no longer
needed or that contain incorrect information. The drop() method might be used to
delete a collection if it is no longer needed or if it needs to be recreated from
scratch.
  Its important to note that deleting documents and collections is a permanent operation 
in MongoDB. Once a document or collection has been deleted, it cannot be recovered.
Therefore, it is important to use these methods with "caution" and to ensure that 
you are deleting the correct documents or collections.


```
